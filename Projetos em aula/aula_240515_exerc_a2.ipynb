```
import numpy as np
import random as rnd
```

## Criar função $znorm = \frac{x - min(x)}{max(x) - min(x)} $

```
def znorm(x):
  
  return (x - min(x)) / (max(x) - min(x))
```

## criar vetores A e B

```
# vetorA = np.arange(1,21) # criando direto com numpy
vetorA = np.array(list(range(1,21))) # metodo apresentado na aula
```

```
vetorA
```

```
# vetorB = np.random.randint(1, 21, 20) # criando direto pelo numpy

vetorB = np.array(rnd.sample(range(1,21), 20)) # metodo apresentado na aula
```

```
vetorB
```

## Aplicar função

```
vetorA_norm = znorm(vetorA)
vetorA_norm
```

```
vetorB_norm = znorm(vetorB)
vetorB_norm
```

## Criar as matrizes

```
np.shape(vetorA), np.shape(vetorB)
```

```
np.shape([vetorA, vetorB])
```

```
np.shape(np.transpose([vetorA, vetorB]))
```

```
matrizAB = np.transpose([vetorA, vetorB])
```

```
matrizAB
```

```
matrizAB_norm = np.transpose([vetorA_norm, vetorB_norm])
```

```
matrizAB_norm
```

## Aplicar operações

- operação *

```
matrizAB*matrizAB_norm
```

- Operação dot

```
matrizAB.dot(np.transpose(matrizAB_norm))
```
