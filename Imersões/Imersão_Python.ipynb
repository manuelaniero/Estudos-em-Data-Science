```python
import pandas as pd
```

Importando os dados

```python
df_principal = pd.read_excel('Imersão Python.xlsx')
```

```python
df_principal
```

```python
df_acoes = pd.read_excel('Imersão Python.xlsx', sheet_name='Total_de_acoes')
df_acoes
```

```python
df_ticker = pd.read_excel('Imersão Python.xlsx', sheet_name='Ticker')
df_ticker
```

```python
df_principal = df_principal[['Ativo', 'Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

Realizando os mesmos calculos feitos na planilha excel

```python
df_principal = df_principal.rename(columns={'Último (R$)': 'valor_final', 'Var. Dia (%)':'valor_dia_per'}).copy()
df_principal
```

```python
df_principal['var_per']= df_principal['valor_dia_per']/100
df_principal['var_inicial']= df_principal['valor_final']/(df_principal['var_per']+1)
df_principal
```

```python
df_principal = df_principal.merge(df_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal['variacao'] = (df_principal['valor_final'] - df_principal['var_inicial']) * df_acoes['Qtde. Teórica']
```

```python
pd.options.display.float_format='{:.2f}'. format
df_principal
```

```python
df_principal['resultado'] =df_principal['variacao'].apply(lambda x :'Subiu' if x>0 else('Desceu' if x < 0 else 'Estável'))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```
